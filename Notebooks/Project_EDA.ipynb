# **Carbon footprint Exploratory Data Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('carbon_footprint_india.csv')
```

## **CO2 emmissions**

```python

sns.barplot(x='Sector', y='CO₂ Emissions (Million Tonnes)', data=df)
plt.title('Distribution of CO2')
plt.xlabel('Sector')
plt.ylabel('CO2 Emission')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='State/UT', y='CO₂ Emissions (Million Tonnes)', data=df)
plt.title('Distribution of CO2 by state')
plt.xlabel('State/UT')
plt.ylabel('CO2 Emission')
plt.show()
```

```python

sns.lineplot(x='Year', y='CO₂ Emissions (Million Tonnes)', data=df)
plt.title('Distribution of CO2 by Year')
plt.xlabel('Year')
plt.ylabel('CO2 Emission')
plt.show()
```

```python
df['Year'].unique()
```

Key Insights:
1. the trend continues to grow
2. by 2025, it is still growing, no signs of reduction
3. not much difference in emmissions across states and sectors

Key Insights:
